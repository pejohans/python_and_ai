### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
#Answer: Let us break down what each object is first:
# 1. Figure = This is the canvas, the drawing surface. It contains one or more Axes. You can set things like title and legend.
# 2. Axes = This is the coordinate system which contains one or more Axis objects
# 3. Axis = Represents x or y axis and its properties like ticks, scale and label
# 4. Artist = This object, the artist is the abstract base class of Figure and Axes.

#So, Figure and Axes are special cases of an Artist object which is used to draw diagrams.
#Axis on the other hand you get from Axes to set for instance the ticks for x or y axis.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
arr = np.linspace(-4, 4, 400)
powerArr = arr**2

# Object Oriented Approach
fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(arr, powerArr, color='blue', linestyle='--')
ax.set_title('Object oriented approach')
ax.set_xlabel('x')
ax.set_ylabel('y')


# pyplot approach
plt.subplots(figsize=(6, 3))
plt.plot(arr, powerArr, color='black', linestyle=':')
plt.title('pyplot approach')
plt.xlabel('x')
plt.ylabel('y')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Scatter figure
fig, axs = plt.subplots(1, 2, figsize=(10, 4), layout='constrained')
axs[0].scatter(random_data_x, random_data_y, s=50, facecolor='C0', edgecolor='k')
axs[0].set_title('Scatter plot')
axs[0].set_ylabel('y')
axs[0].set_xlabel('x')

# Bar figure
axs[1].bar(names, values, color='blue')
axs[1].set_title('Bar plot')
axs[1].set_ylabel('Amount')
axs[1].set_xlabel('Fruits')


```
